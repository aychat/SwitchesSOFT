```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
S0 = 0.05
Delta = 0.03
S1 = np.abs(S0 + Delta*1j)
```

```python
S1
```

```python
T2 = 200.
T1 = 200.
```

```python
t = np.linspace(0., 500., 10000)
```

```python

plt.plot(t, np.exp(-t/T2)*(np.cos(S1*t) + (1./(S1*T2))*np.sin(S1*t)))
plt.show()
```

```python
1./(S1*T2)
```

```python
w0 = -0.9
```

```python

plt.plot(t, w0 - (1 + w0)*np.exp(-t/T2)*(np.cos(S1*t) + (1./(S1*T2))*np.sin(S1*t)))
plt.show()
```
