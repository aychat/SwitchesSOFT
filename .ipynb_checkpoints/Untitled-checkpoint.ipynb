```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
S0 = 1.
Delta = 0.5
S1 = np.abs(S0 + Delta*1j)
```

```python
S1
```

```python
T2 = 200.
T1 = 200.
```

```python
t = np.linspace(0., 1000., 10000)
```

```python

plt.plot(t, np.exp(-t/T2)*(np.cos(S1*t) + (1./(S1*T2))*np.sin(S1*t)))
plt.plot(t, np.exp(-t/T2))
plt.show()
```
